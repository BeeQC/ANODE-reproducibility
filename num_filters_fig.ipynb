```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
losses = [
    {32: [[], [], []],
    64: [[], [], []],
    92: [[], [], []]}
    for _ in range(2)]


for node_type in [0,1]:
    for size in [32, 64, 92]:
        for rep in [0,1,2]:
            try:
                with open(f"results_{size}/losses{node_type}_{rep}.json", "r") as f:
                    losses[node_type][size][rep] = np.array(json.load(f))
            except:
                continue
```

```python
avg_losses = [
    {32: np.mean(losses[i][32],axis=0),
    64:np.mean(losses[i][64],axis=0),
     92:np.mean(losses[i][92],axis=0)}
    
    for i in [0,1]
]
```

```python
total_nfes = [
    {32: [[], [], []],
    64: [[], [], []],
    92: [[], [], []]}
    for _ in range(2)]


for node_type in [0,1]:
    for size in [32, 64, 92]:
        for rep in [0,1,2]:
            try:
                with open(f"results_{size}/total_nfes{node_type}_{rep}.json", "r") as f:
                    total_nfes[node_type][size][rep] = np.array(json.load(f))
            except:
                continue
```

```python
avg_total_nfes = [
    {32: np.mean(total_nfes[i][32],axis=0),
    64:np.mean(total_nfes[i][64],axis=0),
     92:np.mean(total_nfes[i][92][0:2],axis=0)}
    
    for i in [0,1]
]
```

```python
f, axs = plt.subplots(2,2, figsize=(10,10), dpi=300,sharey="row",sharex="row")
plt.rc('font', size=12)

for node_type in [0,1]:

    axs[0,node_type].plot(losses[node_type][32][0], alpha=0.2)
    axs[0,node_type].plot(losses[node_type][32][1], color="#1f77b4", alpha=0.2)
    axs[0,node_type].plot(losses[node_type][32][2], color="#1f77b4", alpha=0.2)

    axs[0,node_type].plot(losses[node_type][64][0], alpha=0.2)
    axs[0,node_type].plot(losses[node_type][64][1], color="#ff5100", alpha=0.2)
    axs[0,node_type].plot(losses[node_type][64][2], color="#ff5100", alpha=0.2)

    axs[0,node_type].plot(losses[node_type][92][0][:len(losses[0][64][0])], alpha=0.2)
    axs[0,node_type].plot(losses[node_type][92][1][:len(losses[0][64][0])], color="#2ca02c", alpha=0.2)
    #axs[0,node_type].plot(losses[node_type][92][2][:len(losses[0][64][0])], color="#2ca02c", alpha=0.2)

    axs[0,node_type].plot(avg_losses[node_type][32], color="#1f77b4", label="Filter Size = 32") 
    axs[0,node_type].plot(avg_losses[node_type][64], color="#ff5100", label="Filter Size = 64")
    axs[0,node_type].plot(avg_losses[node_type][92][:len(losses[0][64][0])], color="#2ca02c", label="Filter Size = 92")
    
    
    
for node_type in [0,1]:

    axs[1,node_type].plot(total_nfes[node_type][32][0], color="#1f77b4", label="Filter Size = 32", alpha=0.2)
    axs[1,node_type].plot(total_nfes[node_type][32][1], color="#1f77b4", alpha=0.2)
    axs[1,node_type].plot(total_nfes[node_type][32][2], color="#1f77b4", alpha=0.2)

    axs[1,node_type].plot(total_nfes[node_type][64][0], color="#ff5100", label="Filter Size = 64", alpha=0.2)
    axs[1,node_type].plot(total_nfes[node_type][64][1], color="#ff5100", alpha=0.2)
    axs[1,node_type].plot(total_nfes[node_type][64][2], color="#ff5100", alpha=0.2)

    axs[1,node_type].plot(total_nfes[node_type][92][0][:len(losses[0][64][0])], color="#2ca02c", label="Filter Size = 92", alpha=0.2)
    axs[1,node_type].plot(total_nfes[node_type][92][1][:len(losses[0][64][0])], color="#2ca02c", alpha=0.2)
    #axs[1,node_type].plot(total_nfes[node_type][92][2][:len(losses[0][64][0])], color="#2ca02c", alpha=0.2)
    
    axs[1,node_type].plot(avg_total_nfes[node_type][32], color="#1f77b4", label="Filter Size = 32") 
    axs[1,node_type].plot(avg_total_nfes[node_type][64], color="#ff5100", label="Filter Size = 64")
    axs[1,node_type].plot(avg_total_nfes[node_type][92][:len(losses[0][64][0])], color="#2ca02c", label="Filter Size = 92")
    



    
axs[0,1].legend()
axs[0,0].set_title("ANODE (p=5)")
axs[0,1].set_title("NODE")    
axs[1,0].set_xlabel("Steps")
axs[1,1].set_xlabel("Steps")
axs[0,0].set_ylabel("Loss")
axs[1,0].set_ylabel("Total NFEs")
        
plt.savefig("filter_size.svg")
```
